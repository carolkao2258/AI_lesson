```python
import pandas as pd
import yfinance as yf

# 輸入檔案路徑
file1 = "o_t51sb11.csv"
file2 = "t51sb11.csv"

# 嘗試用 cp950 編碼讀取 (台灣常用編碼)
df1 = pd.read_csv(file1, encoding="cp950")
df2 = pd.read_csv(file2, encoding="cp950")

# 合併兩個 DataFrame (依列合併)
merged_df = pd.concat([df1, df2], ignore_index=True)

# 篩選出債券型基金 (基金代號最後一個字是 "B")
merged_df = merged_df[merged_df["基金代號"].astype(str).str.endswith("B")].reset_index(drop=True)

# 新增「Previous Close」、「Yield」欄位
merged_df["Previous Close"] = None
merged_df["Yield"] = None

# 根據「基金代號」欄位去抓 Yahoo Finance 的 ETF 資料
for i, row in merged_df.iterrows():
    ticker_symbol = str(row["基金代號"]) + ".TWO"   # 櫃買中心 ETF 要加上 .TWO
    try:
        ticker = yf.Ticker(ticker_symbol)
        info = ticker.info
        
        # 嘗試取得收盤價
        close_price = info.get("previousClose", None)
        if close_price is not None:
            merged_df.at[i, "Previous Close"] = close_price
        else:
            merged_df.at[i, "Previous Close"] = "查無資料"
        
        # 嘗試取得殖利率（已經是百分比格式，不需轉換）
        dividend_yield = info.get("dividendYield", None)
        if dividend_yield is not None:
            merged_df.at[i, "Yield"] = round(dividend_yield, 2)  # 保留兩位小數
        else:
            merged_df.at[i, "Yield"] = "查無資料"
            
    except Exception as e:
        merged_df.at[i, "Previous Close"] = "查無資料"
        merged_df.at[i, "Yield"] = "查無資料"
        print(f"無法取得 {ticker_symbol}: {e}")

# 輸出到新的 CSV 檔案
output_file = "bond_etf_with_close_yield.csv"
```

```python
# 轉換資料格式 (類似 R 的 pivot_longer)
if wa_columns and 'Date' in taiwan_daily_summary.columns:
   wa_data = taiwan_daily_summary.melt(
       id_vars=['Date'],
       value_vars=wa_columns,
       var_name='metric',
       value_name='value'
)
```

## Try it!

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取 CSV
df = pd.read_csv("TWSpotifySentiment.csv", encoding="utf-8-sig")

# 要繪製的欄位
cols = ["wa_valence", "wa_energy", "wa_danceability", "wa_loudness"]

# 找日期欄位
date_like = [c for c in df.columns if any(k in c.lower() for k in
                ["date","datetime","day","time","ts","timestamp"])]
if date_like:
    date_col = date_like[0]
    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
    df = df.sort_values(by=date_col)
    x = df[date_col]
else:
    x = df.index

# 分別繪圖
for col in cols:
    if col in df.columns:
        plt.figure(figsize=(8, 5))
        plt.plot(x, df[col], label=col, color='tab:blue')
        plt.xlabel(date_col if date_like else "Index")
        plt.ylabel("Value")
        plt.title(f"Trend of {col}")
        plt.legend()
        plt.tight_layout()
        plt.savefig(f"{col}_trend.png", dpi=150)
        plt.show()
```

同理，亦可針對avg_columns處理

```python
avg_columns = [col for col in taiwan_daily_summary.columns if col.startswith('avg_')]
print(f"找到 Average 欄位: {avg_columns}")


if avg_columns and 'Date' in taiwan_daily_summary.columns:
   # 轉換資料格式
   avg_data = taiwan_daily_summary.melt(
       id_vars=['Date'],
       value_vars=avg_columns,
       var_name='metric',
       value_name='value'
   )
```

# 練習 9: 使用 Seaborn 製作整合比較

```python
if wa_columns and avg_columns and 'Date' in taiwan_daily_summary.columns:
  # 合併所有資料
  all_data = pd.concat([
  wa_data.assign(type='Weighted Average'),
  avg_data.assign(type='Average')
])

# 使用 seaborn 的 FacetGrid
n_metrics = len(set(wa_columns + avg_columns))
n_cols = 2

g = sns.FacetGrid(all_data, col='metric', hue='type', col_wrap=n_cols,  height=4, aspect=1.2, margin_titles=True)

# 繪製線圖
g.map(plt.plot, 'Date', 'value', marker='o', markersize=2, linewidth=2, alpha=0.8)

# 設定標題和標籤
g.set_axis_labels('Date', 'Value')
g.set_titles('{col_name}')
g.add_legend(title='Type', bbox_to_anchor=(1.05, 1), loc=2)

# 調整 x 軸標籤
for ax in g.axes.flat:
  ax.tick_params(axis='x', rotation=45)
  ax.grid(True, alpha=0.3)

  plt.suptitle('Comparison of Weighted Averages vs Averages of Audio Features in Taiwan', fontsize=16, fontweight='bold', y=1.02)
  plt.tight_layout()
  plt.show()
  print()
```
